```python
import numpy as np
import pandas as pd
```

```python
childsPerWomanOriginal = pd.read_csv('https://raw.githubusercontent.com/MUG594/VIS-Project/main/SourceSheets/children-born-per-woman.csv')
childsPerWomanOriginal = childsPerWomanOriginal.rename(columns = {'Fertility rate (Select Gapminder, v12) (2017)': 'Fertility_Rate' }, inplace = False)
childsPerWomanOriginal
```

```python
continentsCountry = pd.read_csv("https://raw.githubusercontent.com/MUG594/VIS-Project/main/SourceSheets/continents-country.csv")
continentsCountry = continentsCountry.drop(columns=["Entity", "Year"])
continentsCountry
```

```python
countriesGdp = pd.read_csv("https://raw.githubusercontent.com/MUG594/VIS-Project/main/SourceSheets/gdp-capita.csv")
countriesGdp = countriesGdp.drop(columns=["Entity", "145446-annotations"])
countriesGdp = countriesGdp.rename(columns = {'GDP per capita': 'GDP' }, inplace = False)
countriesGdp
```

```python
countriesReligion = pd.read_csv("https://raw.githubusercontent.com/MUG594/VIS-Project/main/SourceSheets/country-religion.csv")
countriesReligion = countriesReligion.drop(columns=["Entity", "Year"])
countriesReligion = countriesReligion.rename(columns = {'Main religion ': 'Religion' }, inplace = False)
countriesReligion
```

```python
countriesSexRatio = pd.read_csv("https://raw.githubusercontent.com/MUG594/VIS-Project/main/SourceSheets/sex-ratio.csv")
countriesSexRatio = countriesSexRatio.drop(columns=["Entity"])
countriesSexRatio = countriesSexRatio.rename(columns = {'Sex ratio at birth (Chao et al. 2019)': 'Sex_Ratio' }, inplace = False)
countriesSexRatio
```

# Merge

```python
resultSet = pd.merge(childsPerWomanOriginal, continentsCountry,"left",left_on="Code", right_on="Code" )
resultSet
```

```python
resultSet = pd.merge(resultSet, countriesGdp,"left",left_on=["Code", "Year"], right_on=["Code", "Year"] )
resultSet
```

```python
resultSet = pd.merge(resultSet, countriesReligion,"left",left_on="Code", right_on="Code" )
resultSet
```

```python
resultSet = pd.merge(resultSet, countriesSexRatio,"left",left_on=["Code", "Year"], right_on=["Code", "Year"] )
resultSet
```

## Cleanup

```python
resultSet.drop_duplicates()
resultSet.drop(resultSet[resultSet.Entity == 'World'].index, inplace=True)
resultSet.drop(resultSet[resultSet.Entity == 'OWID_CIS'].index, inplace=True)
resultSet
```

#Export

```python
resultSet.to_csv("cpw_original.csv", index=False)
```
